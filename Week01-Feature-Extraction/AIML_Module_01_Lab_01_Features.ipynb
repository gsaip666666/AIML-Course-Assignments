We need to clean this up a bit. Let us remove all the special characters and keep only 26 letters and space. Note that this will remove accented characters in French also. We are also removing all the numbers and spaces. So this is not an ideal solution.

```python
def cleanup(text):
  text = text.lower()  # make it lowercase
  text = re.sub('[^a-z]+', '', text) # only keep characters
  return text
```

```python
eng1 = cleanup(eng1)
eng2 = cleanup(eng2)
fr1 = cleanup(fr1)
fr2 = cleanup(fr2)
```

Now let us calculate the frequency of the character n-grams. N-grams are groups of characters of size n. A unigram is a single character and a bigram is a group of two characters and so on.

Let us count the frequency of each character in a text and plot it in a histogram.

```python
# convert a tuple of characters to a string
def tuple2string(tup):
  st = ''
  for ii in tup:
    st = st + ii
  return st

# convert a tuple of tuples to a list of strings
def key2string(keys):
  return [tuple2string(i) for i in keys]

# plot the histogram
def plothistogram(ngram):
  keys = key2string(ngram.keys())
  values = list(ngram.values())

  # sort the keys in alphabetic order
  combined = zip(keys, values)
  zipped_sorted = sorted(combined, key=lambda x: x[0])
  keys, values = map(list, zip(*zipped_sorted))
  plt.bar(keys, values)
```

Let us compare the histograms of English pages and French pages. Can you spot a difference?

```python
unigram_eng1 = Counter(ngrams(eng1,1))
plothistogram(unigram_eng1)
plt.title('English 1')
plt.show()
unigram_eng2 = Counter(ngrams(eng2,1))
plothistogram(unigram_eng2)
plt.title('English 2')
plt.show()
```

```python
unigram_fr1 = Counter(ngrams(fr1,1))
plothistogram(unigram_eng1)
plt.title('French 1')
plt.show()
unigram_fr2 = Counter(ngrams(fr2,1))
plothistogram(unigram_fr2)
plt.title('French 2')
plt.show()
```

We can see that the unigrams for French and English are very similar. So this is not a good feature if we want to distinguish between English and French. Let us look at bigrams.

```python
def plotbihistogram(ngram):
  freq = np.zeros((26,26))
  for ii in range(26):
    for jj in range(26):
      freq[ii,jj] = ngram[(chr(ord('a')+ii), chr(ord('a')+jj))]
  plt.imshow(freq, cmap = 'jet')
  return freq
```

Let us look at the top 10 ngrams for each text.

```python
from IPython.core.debugger import set_trace

def ind2tup(ind):
  ind = int(ind)
  i = int(ind/26)
  j = int(ind%26)
  return (chr(ord('a')+i), chr(ord('a')+j))

def ShowTopN(bifreq, n=10):
  f = bifreq.flatten()
  arg = np.argsort(-f)
  for ii in range(n):
    print(f'{ind2tup(arg[ii])} : {f[arg[ii]]}')
```

**Try With Different Languages**

```python
# Try With Different Languages
from collections import Counter
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

english_text = """
The elephant is the largest land animal. It lives in Africa and Asia.
Elephants are known for their intelligence and memory.
"""

spanish_text = """
El elefante es el animal terrestre más grande. Vive en África y Asia.
Los elefantes son conocidos por su inteligencia y memoria.
"""

french_text = """
Les éléphants sont les plus grands animaux terrestres.
Ils vivent en Afrique et en Asie et sont connus pour leur intelligence.
"""

english_text = english_text.lower()
spanish_text = spanish_text.lower()
french_text = french_text.lower()
```

```python
# Compute Bigram Frequencies for Languages
def bigram_matrix(text):

    alphabet = "abcdefghijklmnopqrstuvwxyz"
    matrix = np.zeros((26,26))

    text = ''.join([c for c in text if c in alphabet])

    for i in range(len(text)-1):
        a = alphabet.index(text[i])
        b = alphabet.index(text[i+1])
        matrix[a][b] += 1

    matrix = matrix / matrix.sum()
    return matrix


eng_bigram = bigram_matrix(english_text)
spa_bigram = bigram_matrix(spanish_text)
fre_bigram = bigram_matrix(french_text)
```

```python
# Visualize Language Bigrams
plt.figure(figsize=(6,5))
sns.heatmap(eng_bigram,cmap="Blues_r")
plt.title("English Bigram Frequencies")
plt.show()
plt.figure(figsize=(6,5))
sns.heatmap(spa_bigram,cmap="mako")
plt.title("Spanish Bigram Frequencies")
plt.show()
plt.figure(figsize=(6,5))
sns.heatmap(fre_bigram)
plt.title("French Bigram Frequencies")
plt.show()

# Observation
# Languages produce different bigram patterns, confirming bigrams help identify language.
```

# Part 2: Written numbers

We will use a subset of the MNIST dataset. Each input character is represented in a 28*28 array. Let us see if we can extract some simple features from these images which can help us distinguish between the digits.

Load the dataset:

```python
from keras.datasets import mnist

#loading the dataset
(train_X, train_y), (test_X, test_y) = mnist.load_data()
```

Extract a subset of the data for our experiment:

```python
no1 = train_X[train_y==1,:,:]
no0 = train_X[train_y==0,:,:]
```

Let us visualize a few images here:

```python
for ii in range(5):
  plt.subplot(1, 5, ii+1)
  plt.imshow(no1[ii,:,:])
plt.show()
for ii in range(5):
  plt.subplot(1, 5, ii+1)
  plt.imshow(no0[ii,:,:])
plt.show()
```

suNow, let us start with a simple feature: the sum of all pixels and see how good this feature is.

```python
sum1 = np.sum(no1>0, (1,2)) # threshold before adding up
sum0 = np.sum(no0>0, (1,2))
```

Let us visualize how good this feature is: (X-axis is mean, y-axis is the digit)

```python
plt.hist(sum1, alpha=0.7);
plt.hist(sum0, alpha=0.7);
```

We can already see that this feature separates the two classes quite well.

Let us look at another, more complicated feature. We will count the number black pixels that are surrounded on four sides by non-black pixels, or "hole pixels".

```python
def cumArray(img):
  img2 = img.copy()
  for ii in range(1, img2.shape[1]):
    img2[ii,:] = img2[ii,:] + img2[ii-1,:]  # for every row, add up all the rows above it.
  img2 = img2>0
  return img2

def getHolePixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3) # rotate and cumulate it again for differnt direction
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4 # this will create a binary image with all the holes filled in.
  hole = hull & ~ (img>0) # remove the original digit to leave behind the holes
  return hole
```

Visualize a few:

```python
imgs = [no1[456,:,:],  no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getHolePixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

Now let us plot the number of hole pixels and see how this feature behaves

```python
hole1 = np.array([getHolePixels(i).sum() for i in no1])
hole0 = np.array([getHolePixels(i).sum() for i in no0])

plt.hist(hole1, alpha=0.7);
plt.hist(hole0, alpha=0.7);
```

This feature works even better to distinguish between one and zero.


Now let us try the number of pixels in the 'hull' or the number with the holes filled in:

```python
def getHullPixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3) # rotate and cumulate it again for differnt direction
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4 # this will create a binary image with all the holes filled in.
  return hull
```

```python
imgs = [no1[456,:,:],  no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getHullPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

Plotting the number of hull pixels versus the digit:

```python
hull1 = np.array([getHullPixels(i).sum() for i in no1])
hull0 = np.array([getHullPixels(i).sum() for i in no0])

plt.hist(hull1, alpha=0.7);
plt.hist(hull0, alpha=0.7);
```

Let us try one more feature, where we look at the number of boundary pixels in each image.

```python
def minus(a, b):
  return a & ~ b

def getBoundaryPixels(img):
  img = img.copy()>0  # binarize the image
  rshift = np.roll(img, 1, 1)
  lshift = np.roll(img, -1 ,1)
  ushift = np.roll(img, -1, 0)
  dshift = np.roll(img, 1, 0)
  boundary = minus(img, rshift) | minus(img, lshift) | minus(img, ushift) | minus(img, dshift)
  return boundary
```

```python
imgs = [no1[456,:,:],  no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getBoundaryPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

```python
bound1 = np.array([getBoundaryPixels(i).sum() for i in no1])
bound0= np.array([getBoundaryPixels(i).sum() for i in no0])

plt.hist(bound1, alpha=0.7);
plt.hist(bound0, alpha=0.7);
```

What will happen if we plot two features together?

**1.** **Sum of Pixels (sum0, sum1)**

This feature represents the total intensity of all pixels in the image. It indicates how much of the image area is filled by the digit. Digits with more strokes or thicker shapes tend to have a higher pixel sum.

**2. Hole Pixels (hole0, hole1)**

This feature counts the number of pixels inside enclosed regions (holes) of the digit. Some digits like 0, 6, 8, and 9 contain holes, while digits like 1 or 7 usually do not.

**3. Convex Hull Pixels (hull0, hull1)**

This feature measures the number of pixels inside the convex hull of the digit, which is the smallest convex shape that completely encloses the digit. It captures the overall outer shape and spread of the digit.

**4. Boundary Pixels (bound0, bound1)**

This feature counts the pixels that lie on the boundary (edges) of the digit. It reflects the outline or perimeter complexity of the digit.

**Feature Combinations**

**1.Which is the best combination of features?**

The best feature combination is the pair that produces the clearest separation between the classes in the scatter plot. In many cases, **sum** **of** **pixels** **vs** **number** **of holes** separates digit 0 and 1 effectively.

**2.How would you test or visualize four or more features?**

When more than two features are present, direct visualization becomes difficult. We can use **pairwise scatter plots** to compare all feature combinations or **3D scatter plots** to visualize three features at once. **3.Can you come up with your own features?**

Yes. Additional features can be designed based on the **structure and shape of the digit**, such as **center pixel density, edge pixel count, vertical symmetry, or horizontal stroke counts.** These features capture different characteristics of the digit and may improve classification performance.

**4.Will these features work for different classes other than 0 and 1?**

These features may work partially but may not be sufficient for all digits. Some digits share similar properties (for example, **3 and 5**, or **0 and 6**), so additional or more discriminative features may be required for accurate classification.

**5.What will happen if we take more than two classes at a time?**

When more classes are included, the **feature distributions of different digits may overlap**, making separation more difficult. This increases the complexity of the classification task and may require **more features or more advanced machine learning models.**

```python
# Pairwise Feature Plot
'''A pairwise feature visualization was created to observe
relationships between multiple features simultaneously.
This helps analyze how well different feature combinations
separate the digit classes.'''

import pandas as pd
import numpy as np
import seaborn as sns

data = pd.DataFrame({
    "sum": np.concatenate((sum0, sum1)),
    "hole": np.concatenate((hole0, hole1)),
    "hull": np.concatenate((hull0, hull1)),
    "boundary": np.concatenate((bound0, bound1)),
    "label": np.concatenate((np.zeros(len(sum0)), np.ones(len(sum1))))
})

sns.pairplot(data, hue="label")
```
